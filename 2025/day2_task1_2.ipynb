## Task 1

```python
import re
```

```python
passwords_list=[]
with open ('day2.txt', 'r', encoding='utf-8') as infile:
    for line in infile:
        passwords_list.extend(line.strip('\n').split(','))
        break

print(passwords_list)
```

```python
ranges_list=[]
for id_range in passwords_list:
    num1, num2= tuple(id_range.split('-'))
    num1=int(num1)
    num2=int(num2)
    x_range = range(num1,num2+1)
    ranges_list.append(x_range)
```

```python
list_repeated_ids = []
for rng in ranges_list:
    for value in rng:
        if len(str(value))%2!=0:
            continue
        else:   
            pat = r"^(\d+)\1$"
            if re.search(pat,str(value)):
                list_repeated_ids.append(value)          
```

```python
results=sum(list_repeated_ids)
```

```python
print(results)
```
