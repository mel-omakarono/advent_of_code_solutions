# Day 5

## Task 1 

```python
task5_input_txt = 'day5.txt'
```

```python
input_ranges =[]
input_ingredients = []
with open (task5_input_txt,'r',encoding='utf-8') as an_input:
    for line in an_input:
        if line.strip('\n').isalnum():
            input_ingredients.append(int(line.strip('\n')))
        else:
            if line.strip('\n')=='':
                continue
            else:
                input_ranges.append(line.strip('\n'))
```

```python
a_set_fresh_ingredients=set()
for a_range in input_ranges:
    rng1,rng2 = a_range.split('-')
    ranges = range(int(rng1),int(rng2)+1)
    for ingredient in input_ingredients:
        if ingredient in ranges:
            a_set_fresh_ingredients.add(ingredient)
```

```python
task1_result = len(a_set_fresh_ingredients)
print(task1_result)
```

## Task 2

```python
list_ranges = []
for a_range in input_ranges:
    if a_range.strip():
        rng1, rng2 = a_range.split('-')
        range_tuple = (int(rng1), int(rng2))
        list_ranges.append(range_tuple)
list_ranges.sort()

total_fresh_count = 0

current_range_start, current_range_end = list_ranges[0]

for next_range_start, next_range_end in list_ranges[1:]:

    if next_range_start <= current_range_end + 1:
        current_range_end = max(current_range_end, next_range_end)

    else:
        total_fresh_count += current_range_end - current_range_start + 1

        current_range_start = next_range_start
        current_range_end = next_range_end

total_fresh_count += current_range_end - current_range_start + 1

print(total_fresh_count)
```
