```python
paper_roll_list = []
with open ('day4.txt','r', encoding='utf-8') as infile:
    for line in infile:
        line = line.strip()
        if line:
            paper_roll_list.append(line)
```

```python
def accessibile_rolls_counter(test_roll):
    accessible_rolls=0
    for line_idx, roll_line in enumerate(test_roll):
        if line_idx == 0:
            for char_idx, roll_char in enumerate(roll_line):
                if roll_char != '@':
                    continue
                else:
                    char_rolls_count=0
                    if char_idx == 0:
                        right_char = roll_line[char_idx + 1]
                        bottom_char = test_roll[line_idx + 1][char_idx]
                        bottom_right = test_roll[line_idx + 1][char_idx+1]
                        accessible_rolls+=1
        
                    elif char_idx<len(roll_line)-1:
                        left_char= roll_line[char_idx - 1]
                        right_char = roll_line[char_idx + 1]
                        bottom_char = test_roll[line_idx + 1][char_idx]
                        bottom_left=test_roll[line_idx + 1][char_idx-1]
                        bottom_right = test_roll[line_idx + 1][char_idx+1]
                        roll_surroundings=[left_char,right_char,bottom_char,bottom_left,bottom_right]
                        for roll_surr in roll_surroundings:
                            if roll_surr=='@':
                                char_rolls_count+=1
                        if char_rolls_count>=4:
                            continue
                        else:
                            accessible_rolls+=1
                            
                    else:
                        left_char= roll_line[char_idx - 1]
                        bottom_char = test_roll[line_idx + 1][char_idx]
                        bottom_left=test_roll[line_idx + 1][char_idx-1]
                        accessible_rolls+=1
        
        elif line_idx<len(roll_line)-1:
            for char_idx, roll_char in enumerate(roll_line):
                if roll_char != '@':
                    continue
                else:
                    char_rolls_count=0
                    if char_idx == 0:
                        up_char=test_roll[line_idx - 1][char_idx]
                        up_right = test_roll[line_idx - 1][char_idx+1]
                        right_char = roll_line[char_idx + 1]
                        bottom_char = test_roll[line_idx + 1][char_idx]
                        bottom_right = test_roll[line_idx + 1][char_idx+1]
                        roll_surroundings=[up_char,up_right,right_char,bottom_char,bottom_right]
                        for roll_surr in roll_surroundings:
                            if roll_surr=='@':
                                char_rolls_count+=1
                        if char_rolls_count>=4:
                            continue
                        else:
                            accessible_rolls+=1
        
                    elif char_idx<len(roll_line)-1:
                        up_char = test_roll[line_idx - 1][char_idx]
                        up_left=test_roll[line_idx - 1][char_idx-1]
                        up_right = test_roll[line_idx - 1][char_idx+1]
                        left_char= roll_line[char_idx - 1]
                        right_char = roll_line[char_idx + 1]
                        bottom_char = test_roll[line_idx + 1][char_idx]
                        bottom_left=test_roll[line_idx + 1][char_idx-1]
                        bottom_right = test_roll[line_idx + 1][char_idx+1]
    
                        roll_surroundings=[up_char,up_left,up_right,left_char,right_char,bottom_left,bottom_char,bottom_right]
                        for roll_surr in roll_surroundings:
                            if roll_surr=='@':
                                char_rolls_count+=1
                        if char_rolls_count>=4:
                            continue
                        else:
                            accessible_rolls+=1
                    else:
                        up_char = test_roll[line_idx - 1][char_idx]
                        up_left=test_roll[line_idx - 1][char_idx-1]
                        left_char= roll_line[char_idx - 1]
                        bottom_char = test_roll[line_idx + 1][char_idx]
                        bottom_left=test_roll[line_idx + 1][char_idx-1]
    
                        roll_surroundings=[up_char,up_left,left_char,bottom_left,bottom_char]
                        for roll_surr in roll_surroundings:
                            if roll_surr=='@':
                                char_rolls_count+=1
                        if char_rolls_count>=4:
                            continue
                        else:
                            accessible_rolls+=1
        else:
            for char_idx, roll_char in enumerate(roll_line):
                if roll_char != '@':
                    continue
                else:
                    char_rolls_count=0
                    if char_idx == 0:
                        right_char = roll_line[char_idx + 1]
                        up_char = test_roll[line_idx - 1][char_idx]
                        up_right = test_roll[line_idx - 1][char_idx+1]
                        accessible_rolls+=1
            
                    elif char_idx<len(roll_line)-1:
                        left_char= roll_line[char_idx - 1]
                        right_char = roll_line[char_idx + 1]
                        up_char = test_roll[line_idx - 1][char_idx]
                        up_left=test_roll[line_idx - 1][char_idx-1]
                        up_right = test_roll[line_idx - 1][char_idx+1]
    
                        roll_surroundings=[up_char,up_left,up_right,left_char,right_char]
                        for roll_surr in roll_surroundings:
                            if roll_surr=='@':
                                char_rolls_count+=1
                        if char_rolls_count>=4:
                            continue
                        else:
                            accessible_rolls+=1
                    else:
                        left_char= roll_line[char_idx - 1]
                        up_char = test_roll[line_idx - 1][char_idx]
                        up_left=test_roll[line_idx - 1][char_idx-1]
                        accessible_rolls+=1
                        
                
    return accessible_rolls
```

```python
def removable_rolls_counter(test_roll):
    total_removed = 0
    something_removed = True
    
    grid = list(test_roll)  
    
    while something_removed:
        something_removed = False
        accessible_rolls = 0
        to_remove = []
        
        for line_idx, roll_line in enumerate(grid):
            if line_idx == 0:  
                for char_idx, roll_char in enumerate(roll_line):
                    if roll_char != '@':
                        continue
                    else:
                        char_rolls_count = 0
                        if char_idx == 0:  
                            to_remove.append((line_idx, char_idx))
                        
                        elif char_idx < len(roll_line) - 1: 
                            left_char = roll_line[char_idx - 1]
                            right_char = roll_line[char_idx + 1]
                            bottom_char = grid[line_idx + 1][char_idx]
                            bottom_left = grid[line_idx + 1][char_idx - 1]
                            bottom_right = grid[line_idx + 1][char_idx + 1]
                            
                            roll_surroundings = [left_char, right_char, bottom_char, bottom_left, bottom_right]
                            for roll_surr in roll_surroundings:
                                if roll_surr == '@':
                                    char_rolls_count += 1
                            if char_rolls_count < 4:
                                to_remove.append((line_idx, char_idx))
                        
                        else:  
                            to_remove.append((line_idx, char_idx))
            
            elif line_idx < len(grid) - 1: 
                for char_idx, roll_char in enumerate(roll_line):
                    if roll_char != '@':
                        continue
                    else:
                        char_rolls_count = 0
                        if char_idx == 0:  
                            up_char = grid[line_idx - 1][char_idx]
                            up_right = grid[line_idx - 1][char_idx + 1]
                            right_char = roll_line[char_idx + 1]
                            bottom_char = grid[line_idx + 1][char_idx]
                            bottom_right = grid[line_idx + 1][char_idx + 1]
                            
                            roll_surroundings = [up_char, up_right, right_char, bottom_char, bottom_right]
                            for roll_surr in roll_surroundings:
                                if roll_surr == '@':
                                    char_rolls_count += 1
                            if char_rolls_count < 4:
                                to_remove.append((line_idx, char_idx))
                        
                        elif char_idx < len(roll_line) - 1:  
                            up_char = grid[line_idx - 1][char_idx]
                            up_left = grid[line_idx - 1][char_idx - 1]
                            up_right = grid[line_idx - 1][char_idx + 1]
                            left_char = roll_line[char_idx - 1]
                            right_char = roll_line[char_idx + 1]
                            bottom_char = grid[line_idx + 1][char_idx]
                            bottom_left = grid[line_idx + 1][char_idx - 1]
                            bottom_right = grid[line_idx + 1][char_idx + 1]
                            
                            roll_surroundings = [up_char, up_left, up_right, left_char, right_char, 
                                                bottom_left, bottom_char, bottom_right]
                            for roll_surr in roll_surroundings:
                                if roll_surr == '@':
                                    char_rolls_count += 1
                            if char_rolls_count < 4:
                                to_remove.append((line_idx, char_idx))
                        
                        else: 
                            up_char = grid[line_idx - 1][char_idx]
                            up_left = grid[line_idx - 1][char_idx - 1]
                            left_char = roll_line[char_idx - 1]
                            bottom_char = grid[line_idx + 1][char_idx]
                            bottom_left = grid[line_idx + 1][char_idx - 1]
                            
                            roll_surroundings = [up_char, up_left, left_char, bottom_left, bottom_char]
                            for roll_surr in roll_surroundings:
                                if roll_surr == '@':
                                    char_rolls_count += 1
                            if char_rolls_count < 4:
                                to_remove.append((line_idx, char_idx))
            
            else:  
                for char_idx, roll_char in enumerate(roll_line):
                    if roll_char != '@':
                        continue
                    else:
                        char_rolls_count = 0
                        if char_idx == 0:  
                            to_remove.append((line_idx, char_idx))
                        
                        elif char_idx < len(roll_line) - 1:  
                            left_char = roll_line[char_idx - 1]
                            right_char = roll_line[char_idx + 1]
                            up_char = grid[line_idx - 1][char_idx]
                            up_left = grid[line_idx - 1][char_idx - 1]
                            up_right = grid[line_idx - 1][char_idx + 1]
                            
                            roll_surroundings = [up_char, up_left, up_right, left_char, right_char]
                            for roll_surr in roll_surroundings:
                                if roll_surr == '@':
                                    char_rolls_count += 1
                            if char_rolls_count < 4:
                                to_remove.append((line_idx, char_idx))
                        
                        else: 
                            to_remove.append((line_idx, char_idx))
        
        if to_remove:
            something_removed = True
            total_removed += len(to_remove)
            
            grid = [list(row) for row in grid]
            
            for line_idx, char_idx in to_remove:
                grid[line_idx][char_idx] = '.'
            
            grid = [''.join(row) for row in grid]
    
    return total_removed

result_task2 = removable_rolls_counter(paper_roll_list)
print(result_task2)
```
