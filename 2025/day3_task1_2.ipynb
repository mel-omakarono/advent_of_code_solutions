## Task 1

```python
joltage_list=[]
with open ('day3.txt', 'r', encoding = 'utf-8') as infile:
    for line in infile:
        line = line.strip()
        if line:
            joltage_list.append(line)
        
```

```python
print(len(joltage_list))
```

```python
test_list = ['987654321111111', '811111111111119', '234234234234278', '818181911112111']
```

```python
list_digits = []
for jolt in joltage_list:
    list_jolts_digit=[]
    for idx, digit in enumerate(jolt):
        list_jolts_digit.append(digit)
    highest_digit = max(list_jolts_digit[:-1])
    highest_digit_idx=list_jolts_digit.index(highest_digit)
    second_highest= max(list_jolts_digit[highest_digit_idx+1:])
    new_string = str(highest_digit)+str(second_highest)
    list_digits.append(int(new_string))

results_task1 = sum(list_digits)
```

```python
print(results_task1)
```

## Task 2

```python
list_digits = []

for jolt in joltage_list:
    digits = list(jolt)
    result = []
    idx = 0

    total_len = len(digits)
    target_len = 12

    while len(result) < target_len:
        remaining_needed = target_len - len(result)

        search_end = total_len - remaining_needed + 1

        highest_digit = max(digits[idx:search_end])
        highest_idx = digits.index(highest_digit, idx, search_end)

        result.append(highest_digit)
        idx = highest_idx + 1

    list_digits.append(int("".join(result)))

results_task2 = sum(list_digits)
print(results_task2)
```
